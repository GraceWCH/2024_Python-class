```python
#function有傳出值
import random
def getRandomNum()->float:
    return random.random()

n:float = getRandomNum()
print(n)
```

```python
#一個參數
def sayHello(n:str)->None:
    name:str = n
    print(f"Hello!{name}")

sayHello("Robert") #引數值
sayHello(n="徐")   #引數名稱的呼叫
```

```python
#2個參數
#傳出值int

def multiply(a:int,b:int) -> int:
    return a * b

print(multiply(5,10))         #引數值的呼叫
print(multiply(a=5,b=10))     #引數名稱的呼叫
```
