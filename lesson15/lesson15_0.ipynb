```python
import matplotlib.pyplot as plt

labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [20, 30, 45, 10]
colors = ['yellow', 'green', 'red', 'blue']

fig = plt.figure(figsize=(4.9,4.15))
axes = fig.add_subplot()
axes.pie(values, 
         explode = (0.3, 0, 0, 0), 
         labels = labels, 
         colors = colors, 
         autopct = "%1.2f%%",
         startangle = 180.0,
         shadow = True,
         radius = 1,
         center = (0, 0)
         )
plt.show()
```
