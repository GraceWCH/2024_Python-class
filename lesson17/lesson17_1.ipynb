```python
import numpy as np
import pandas as pd
students = np.random.randint(50,101,size=(50,5))
stu = pd.DataFrame(students,
             columns=['國文','英文','數學','地理','歷史'],
             index=[f'第{i}號' for i in range(1,51)]
             )
stu
```

```python
#陣列運算,使用數學運算子
stu['國文'] + stu['數學']
```

```python
#陣列運算,使用比較運算子
stu['國文'] <= 60
```

```python
#Series執行統計的方法
stu['國文'].sum()
```

```python
#DataFrame執行統計的方法
stu
stu['總分'] = stu.iloc[:,:5].sum(axis=1)
stu['平均'] = stu.iloc[:,:5].mean(axis=1)
stu
```

```python
stu
stu['總分'] = np.sum(stu.iloc[:,:5],axis=1)
stu['平均'] = np.mean(stu.iloc[:,:5],axis=1)
stu
```

```python
#使用numpy統計方法
import numpy as np
np.sum(stu["國文"])
```

```python
#DataFrame執行統計的方法
stu
stu["總分"] = stu.iloc[:,0:5].sum(axis=1)     #[0:5]可省略寫為[:5]
stu["平均"] = stu.iloc[:,0:5].mean(axis=1)
stu
```

```python
stu
stu["總分"] = np.sum(stu.iloc[:,0:5],axis=1)     
stu["平均"] = np.mean(stu.iloc[:,0:5],axis=1)  
stu
```

```python
#使用series的索引和index name
stu["總分"].iloc[0]
stu["總分"]["第1號"]
```

```python
#sort
stu["總分"].sort_values()
sorted_s = stu["總分"].sort_values(ascending=False)
```

```python
#使用排序sort_values(),取出前5名
sorted_s[:5]
```

```python
#DataFrame sorted
sorted_df = stu.sort_values(by='總分',ascending=False)
sorted_df[:5]
```

```python
sorted_df["排名"] = sorted_df["平均"].rank(ascending=False,method="first")
sorted_df
```

```python
#國文不及格的
sorted_df[sorted_df["國文"] < 60]
```

```python
#使用subscript, 搜尋國文及格同時英文及格
sorted_df[(sorted_df["國文"] >= 60) & (sorted_df["英文"] >= 60)]
import numpy as np
sorted_df[(sorted_df["國文"] >= 60) & (sorted_df["英文"] >= 60)].shape[0]
```
